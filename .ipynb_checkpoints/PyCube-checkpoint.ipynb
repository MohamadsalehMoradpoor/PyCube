```python
from sqlalchemy import create_engine
from cubes.tutorial.sql import create_table_from_csv
```

```python
engine = create_engine('sqlite:///data7learnstudents.sqlite')
```

```python
create_table_from_csv(engine,
                     'data.csv',
                     table_name='saling_data',
                     fields=[
                         ('category', 'string'),
                         ('category_label', 'string'),
                         ('subcategory', 'string'),
                         ('subcategory_label', 'string'),
                         ('line_item', 'string'),
                         ('year', 'integer'),
                         ('amount', 'integer')
                     ],
                     create_id=True)
```

```python
from cubes import Workspace, Cell, PointCut
```

```python
ws = Workspace()
ws.register_default_store('sql', url='sqlite:///data7learnstudents.sqlite')
ws.import_model('model.json')
```

```python
b = ws.browser('saling_data')
```

```python
result = b.aggregate()
```

```python
result.summary['amount_sum']
```

```python
result = b.aggregate(drilldown=['item'])
```

```python
for row in result.table_rows('item'):
    print(row.label)
    print(row.record['record_count'])
    print(row.record['amount_sum'])
    print(row.record['double_amount_sum'])
```
